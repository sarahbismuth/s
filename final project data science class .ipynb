# Domain knowledge and definition of the variables 

age:In general people that are 65 and/or older are more likely to have heart issues(https://wa.kaiserpermanente.org/healthAndWellness/index.jhtml?item=/common/healthAndWellness/conditions/heartDisease/ageAndGender.html)

anaemia:  is a decrease in the total amount of red blood cells (RBCs) or hemoglobin in the blood, or a lowered ability of the blood to carry oxygen (https://en.wikipedia.org/wiki/Anemia#:~:text=Anemia%20(also%20spelled%20anaemia)%20is,a%20poor%20ability%20to%20exercise)

creatinine_phosphokinase  (CPK): is an enzyme  expressed by various tissues and cell types,here the numbers represent  the Level of the CPK enzyme in the blood (mcg/L).A normal CPK value is 10 - 120 micrograms per liter.(https://www.ucsfbenioffchildrens.org/tests/003503.html)


diabetes: 0 if the patient does not have diabetes, 1 if the patient has diabetes (to see what is diabetes: https://www.cdc.gov/diabetes/basics/diabetes.html#:~:text=Diabetes%20is%20a%20chronic%20long,your%20pancreas%20to%20release%20insulin.)

ejection_fraction:Percentage of blood leaving the heart at each contraction. An ejection fraction of 55 percent or higher is considered normal.

high_blood_pressure: boolean variable, 1 If the patient has hypertension, 0 if he has not.(https://www.mayoclinic.org/diseases-conditions/high-blood-pressure/symptoms-causes/syc-20373410#:~:text=High%20blood%20pressure%20is%20a,problems%2C%20such%20as%20heart%20disease)

platelets: Platelets in the blood .A normal platelet count ranges from 150,000 to 450,000 platelets per microliter of blood.

serum_creatinine: Level of serum creatinine in the blood (mg/dL).Creatinine is a waste product that comes from the normal wear and tear on muscles of the body.Normal levels of creatinine in the blood are approximately 0.6 to 1.2 milligrams (mg) per deciliter (dL) in adult males and 0.5 to 1.1 milligrams per deciliter in adult females.

serum_sodium:Level of serum sodium in the blood (mEq/L).The normal range for blood sodium levels is 135 to 145 milliequivalents per liter (mEq/L)

sex:Woman(0) or man(1)

smoking:If the patient smokes or not (boolean)

time:Follow-up period (days)

DEATH_EVENT:If the patient deceased during the follow-up period (1 if yes 0 if no)

# Feature selection 

We are now going to use the forward selection as a feature selection method, to select the variables that will make the best prediction.

```python
#forward selection 
#We are first going to put k_features= 12 IN ORDER TO have a general idea of how many variables will lead to good performance
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
from sklearn.linear_model import LinearRegression
# Sequential Forward Selection(sfs)
sfs = SFS(LinearRegression(),
           k_features=12,
           forward=True,
           floating=False,
           scoring = 'r2',
           cv = 0)
X = df.drop(["DEATH_EVENT"],1)
y = df["DEATH_EVENT"]
sfs.fit(X, y)
fig1 = plot_sfs(sfs.get_metric_dict(), kind='std_dev')
plt.title('Sequential Forward Selection (w. StdErr)')
plt.grid()
```

As we can see 7 seems to be a good number of variables to use to have the performance. Indeed, after 7 variables using more variables does not leed to better performance. Thus for our model, we are going to use 7 variables.

We are now going to re-run this code and put K_feature=7, to see the name of these 7 variables that give the best performance.

```python
#forward selection 
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
from sklearn.linear_model import LinearRegression
# Sequential Forward Selection(sfs)
sfs = SFS(LinearRegression(),
           k_features=7,
           forward=True,
           floating=False,
           scoring = 'r2',
           cv = 0)
X = df.drop(["DEATH_EVENT"],1)
y = df["DEATH_EVENT"]
sfs.fit(X, y)
fig1 = plot_sfs(sfs.get_metric_dict(), kind='std_dev')
plt.title('Sequential Forward Selection (w. StdErr)')
plt.grid()
sfs.k_feature_names_
```

# Models

```python
Before to construct our models we are going to split the data.
```

```python
#Define X and Y
X1=df2.drop(["DEATH_EVENT"],1)
Y1=df2['DEATH_EVENT']
```

```python
#split the data 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    X1, Y1, test_size=0.3, random_state=10)
```

Now that we have split our data , we are going to construct 4 different models.

We are going to start with decision tree model.

```python
#import the necessary library 
from sklearn.tree import DecisionTreeClassifier
```

```python
#decision tree model
clf = DecisionTreeClassifier(criterion = 'entropy')
```

I have decided to had the entropy  critarion since it attempts to maximize the mutual information  in the decision tree

```python
#Training the decision tree classifier. 
clf.fit(x_train, y_train)
```

```python
#Predicting labels on the test set.
y_pred_tree=  clf.predict(x_test)
```

```python
#Importing the accuracy metric from sklearn.metrics library

from sklearn.metrics import accuracy_score
print('Accuracy Score on train data: ', accuracy_score(y_true=y_train, y_pred=clf.predict(x_train)))
print('Accuracy Score on test data: ', accuracy_score(y_true=y_test, y_pred=y_pred_tree))
```

Looking at the accuracy on the train data, we can see that the model is overfitted, this can be since our response variable is not balanced.We thus need to look at other metric to evaluate the performance of our model.

```python
from sklearn.metrics import classification_report
print('\nClassification Report\n')
print(classification_report(y_test, y_pred_tree, target_names=['0','1']))
```

We are also going to look at the confusion matrix 

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred_tree)
```
